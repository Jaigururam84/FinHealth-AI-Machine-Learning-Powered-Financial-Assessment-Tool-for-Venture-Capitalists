```python
# Function to detect and cap outliers using IQR
def cap_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Cap the outliers
        df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
        df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    return df

# Identify numeric columns
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns

# Apply the outlier capping function
data = cap_outliers(data, numeric_cols)
```

```python
#checking outlier are treated Okay
from scipy import stats
for col in numeric_cols:
    Q1 = np.percentile(data[col], 25, interpolation='midpoint')
    Q3 = np.percentile(data[col], 75, interpolation='midpoint')
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[~((data[col] >= lower_bound) & (data[col] <= upper_bound))]

# Print information about outliers
    print(f"Outliers for column '{col}':")
    print(f"  Number of outliers: {len(outliers)}")
    print(f"  Outlier values: {outliers[col].to_list()}")
```

```python
# Creating the Target Variable
# The target variable is default and should take the value 1 when net worth next year is negative & 0 when net worth next year is positive
data['default'] = (data['Networth Next Year'] <= 0).astype(int)
```

```python
# Data split
from sklearn.model_selection import train_test_split

# Define features and target
X = data.drop(columns=['Networth Next Year', 'default', 'Num'])  # Drop irrelevant/target columns
y = data[['default']]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Scale the Data
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Model Building
# Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, roc_curve

# Initialize and train Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)

# Predictions
y_pred_lr = log_reg.predict(X_test_scaled)
y_proba_lr = log_reg.predict_proba(X_test_scaled)[:, 1]

# Metrics
print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr))
print(f"AUC-ROC: {roc_auc_score(y_test, y_proba_lr)}")

# Plot ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_proba_lr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"Logistic Regression (AUC = {roc_auc_score(y_test, y_proba_lr):.2f})")
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
# Random forest 
from sklearn.ensemble import RandomForestClassifier

# Initialize and train Random Forest
rf = RandomForestClassifier(random_state=42, n_estimators=100)
rf.fit(X_train_scaled, y_train)

# Predictions
y_pred_rf = rf.predict(X_test_scaled)
y_proba_rf = rf.predict_proba(X_test_scaled)[:, 1]

# Metrics
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
print(f"AUC-ROC: {roc_auc_score(y_test, y_proba_rf)}")

# Plot ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_proba_rf)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"Random Forest (AUC = {roc_auc_score(y_test, y_proba_rf):.2f})")
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Confusion Matrix for Logistic Regression
cm_lr = confusion_matrix(y_test, y_pred_lr)
disp_lr = ConfusionMatrixDisplay(confusion_matrix=cm_lr, display_labels=["Non-Default", "Default"])
disp_lr.plot(cmap='Blues')
plt.title("Logistic Regression Confusion Matrix")
plt.show()

# Confusion Matrix for Random Forest
cm_rf = confusion_matrix(y_test, y_pred_rf)
disp_rf = ConfusionMatrixDisplay(confusion_matrix=cm_rf, display_labels=["Non-Default", "Default"])
disp_rf.plot(cmap='Blues')
plt.title("Random Forest Confusion Matrix")
plt.show()
```

```python
# Dealing Multicollinearity
# VIF threshold i am going to consider is 10 as with a large number of variables, using a stricter threshold of 5 might lead to removing too many variable
from statsmodels.stats.outliers_influence import variance_inflation_factor
X = data.drop(columns=['Networth Next Year', 'default', 'Num'])  # Drop irrelevant/target columns

# Calculate VIF for each predictor
def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

# Looping VIF and columns dropping 
while True:
    vif = calculate_vif(X)
    max_vif = vif['VIF'].max()
    
    if max_vif < 5: # Threshold as 5 
        break
    
    # Find the feature with the highest VIF
    feature_to_drop = vif.loc[vif['VIF'] == max_vif, 'Feature'].iloc[0]
    
    print(f"Dropping {feature_to_drop} with VIF {max_vif:.2f}")
    
    # Drop the feature with the highest VIF
    X = X.drop(columns=[feature_to_drop])

print("\nRemaining features:")
print(X.columns.tolist())

print("\nFinal VIF values:")
print(calculate_vif(X))
```

```python
X_final = X
y_final = data[['default']]

X_train_final, X_test_final, y_train_final, y_test_final = train_test_split(X_final, y_final, test_size=0.2, random_state=1, stratify=y)
print(X_train_final.shape, X_test_final.shape, y_train_final.shape, y_test_final.shape)
```

```python
# Logistic Regression model after removing the Multicollinearity
logistic_model_1 = LogisticRegression(max_iter=1000)
logistic_model_1.fit(X_train_final, y_train_final)
# Make predictions
y_pred_logistic_1 = logistic_model_1.predict(X_test_final)
```

```python
# Random Forest model after removing the Multicollinearity
rf_model_1 = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model_1.fit(X_train_final, y_train_final)

# Make predictions
y_pred_rf_1= rf_model_1.predict(X_test_final)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# Evaluate Logistic Regression performance
accuracy_logistic = accuracy_score(y_test_final, y_pred_logistic_1)
precision_logistic = precision_score(y_test_final, y_pred_logistic_1)
recall_logistic = recall_score(y_test_final, y_pred_logistic_1)
f1_logistic = f1_score(y_test_final, y_pred_logistic_1)
roc_auc_logistic = roc_auc_score(y_test_final, logistic_model_1.predict_proba(X_test_final)[:, 1])

print("Logistic Regression Performance:")
print(f"Accuracy: {accuracy_logistic:.4f}")
print(f"Precision: {precision_logistic:.4f}")
print(f"Recall: {recall_logistic:.4f}")
print(f"F1 Score: {f1_logistic:.4f}")
print(f"ROC-AUC: {roc_auc_logistic:.4f}")

# Evaluate Random Forest performance
accuracy_rf = accuracy_score(y_test_final, y_pred_rf_1)
precision_rf = precision_score(y_test_final, y_pred_rf_1)
recall_rf = recall_score(y_test_final, y_pred_rf_1)
f1_rf = f1_score(y_test_final, y_pred_rf_1)
roc_auc_rf = roc_auc_score(y_test_final, rf_model_1.predict_proba(X_test_final)[:, 1])

print("\nRandom Forest Performance:")
print(f"Accuracy: {accuracy_rf:.4f}")
print(f"Precision: {precision_rf:.4f}")
print(f"Recall: {recall_rf:.4f}")
print(f"F1 Score: {f1_rf:.4f}")
print(f"ROC-AUC: {roc_auc_rf:.4f}")
```

```python
# Optimal Thersold for Logistic regression model
from sklearn.metrics import auc
# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test_final, logistic_model_1.predict_proba(X_test_final)[:, 1])
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='blue', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid()
plt.show()

# Determine optimal threshold based on Youden's J statistic
optimal_idx = np.argmax(tpr - fpr)
optimal_threshold = thresholds[optimal_idx]
print(f'Optimal Threshold: {optimal_threshold:.2f}')
```

```python
# Hyperparameter Tuning for Random Forest 
from sklearn.model_selection import RandomizedSearchCV

# Define hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [None, 10, 20, 30, 40],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

# Random Forest Model
rf = RandomForestClassifier(random_state=42)

# Randomized Search
rf_random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_grid, 
                                      n_iter=50, scoring='roc_auc', cv=5, verbose=2, 
                                      random_state=42, n_jobs=-1)
rf_random_search.fit(X_train_final, y_train_final)

# Best parameters and performance
print("Best Parameters:", rf_random_search.best_params_)
print("Best AUC Score:", rf_random_search.best_score_)

# Final Model Evaluation
best_rf = rf_random_search.best_estimator_
y_rf_probs = best_rf.predict_proba(X_test_final)[:, 1]

print("AUC-ROC:", roc_auc_score(y_test_final, y_rf_probs))
print("Confusion Matrix:\n", confusion_matrix(y_test_final, best_rf.predict(X_test_final)))
print("Classification Report:\n", classification_report(y_test_final, best_rf.predict(X_test_final)))
```

```python
# Model performance check across different metrics
# Evaluate Logistic Regression Model
# Use the optimal threshold for predictions
y_pred_logistic_optimal = (logistic_model_1.predict_proba(X_test_final)[:, 1] >= optimal_threshold).astype(int)

# Logistic Regression Metrics
logistic_accuracy = accuracy_score(y_test_final, y_pred_logistic_optimal)
logistic_precision = precision_score(y_test_final, y_pred_logistic_optimal)
logistic_recall = recall_score(y_test_final, y_pred_logistic_optimal)
logistic_f1 = f1_score(y_test_final, y_pred_logistic_optimal)
logistic_auc = roc_auc_score(y_test_final, logistic_model_1.predict_proba(X_test_final)[:, 1])
logistic_cm = confusion_matrix(y_test_final, y_pred_logistic_optimal)

print("Logistic Regression Model Performance:")
print(f"Accuracy: {logistic_accuracy:.2f}")
print(f"Precision: {logistic_precision:.2f}")
print(f"Recall: {logistic_recall:.2f}")
print(f"F1-Score: {logistic_f1:.2f}")
print(f"AUC-ROC: {logistic_auc:.2f}")
print("Confusion Matrix:\n", logistic_cm)
print("Classification Report:\n", classification_report(y_test_final, y_pred_logistic_optimal))


# Evaluate Random Forest Model
# Use the predictions from the best Random Forest model after hyperparameter tuning
y_rf_probs_best = best_rf.predict_proba(X_test_final)[:, 1]
y_pred_rf_best = best_rf.predict(X_test_final)

# Random Forest Metrics
rf_accuracy = accuracy_score(y_test_final, y_pred_rf_best)
rf_precision = precision_score(y_test_final, y_pred_rf_best)
rf_recall = recall_score(y_test_final, y_pred_rf_best)
rf_f1 = f1_score(y_test_final, y_pred_rf_best)
rf_auc = roc_auc_score(y_test_final, y_rf_probs_best)
rf_cm = confusion_matrix(y_test_final, y_pred_rf_best)

print("\nRandom Forest Model Performance:")
print(f"Accuracy: {rf_accuracy:.2f}")
print(f"Precision: {rf_precision:.2f}")
print(f"Recall: {rf_recall:.2f}")
print(f"F1-Score: {rf_f1:.2f}")
print(f"AUC-ROC: {rf_auc:.2f}")
print("Confusion Matrix:\n", rf_cm)
print("Classification Report:\n", classification_report(y_test_final, y_pred_rf_best))


# Comparison Summary
print("\nComparison Summary:")
print(f"Logistic Regression -> Accuracy: {logistic_accuracy:.2f}, AUC-ROC: {logistic_auc:.2f}")
print(f"Random Forest -> Accuracy: {rf_accuracy:.2f}, AUC-ROC: {rf_auc:.2f}")
```
